```python
train_SF = train_SF.dropna()
```

```python
sf_train, sf_test = train_SF.random_split(.6, seed=5)
```

```python
del sf_train['campaign_id']
del sf_test['campaign_id']
```

```python
sf_train.show()
```

```python
sf_test.show()
```

```python
from graphlab.toolkits.feature_engineering import OneHotEncoder
encoder = gl.feature_engineering.create(sf_train,
                    OneHotEncoder(excluded_features=['X1','display_id','ad_id','clicked']))
transformed_train_sf = encoder.transform(sf_train)
transformed_train_sf.head()
```

```python
transformed_test_sf = encoder.transform(sf_test)
transformed_test_sf.head()
```

```python
model = gl.classifier.create(transformed_train_sf, target='clicked',
                             features = [])
predictions = model.classify(transformed_test_sf)
# Evaluate the model, with the results stored in a dictionary
results = model.evaluate(transformed_test_sf)
```
